```python
import pandas as pd

average_portfo_values = [500000000, 
                         1000000000, 
                         2500000000, 
                         5000000000]
prediction_thresholds = [5000000000, 
                         10000000000, 
                         15000000000, 
                         30000000000]


# Initialize an empty dictionary to store results
result_matrix = {}

# Iterate through each combination
for prediction in prediction_thresholds:
    for portfo_value in average_portfo_values:

        merged_df['portfo'] = merged_df['average'].apply(lambda x: 'low' if x < portfo_value else 'High')    
        merged_df['house'] = merged_df['estimated_price'].apply(lambda x: 'low' if x < prediction else 'High')
        fisher_table = merged_df[['house','portfo']]

        contingency_table = pd.crosstab(fisher_table['portfo'], fisher_table['house'])

        # Calculate total number of observations
        total_obs = contingency_table.values.sum()

        # Calculate probabilities for each cell
        probabilities = contingency_table / total_obs

        x = probabilities.iloc[0,0]/(probabilities.iloc[0,0]+probabilities.iloc[1,0])
        x = round(x,3)*100

        # Store the result in the dictionary
        result_matrix[(prediction, portfo_value)] = x

# Create a DataFrame from the dictionary
result_df = pd.DataFrame(result_matrix, index=['x']).T
result_df.index.names = ['Prediction Threshold', 'Average Portfolio']

# Print the result
result_df
```
