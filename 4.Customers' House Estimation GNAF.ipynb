### Loading Data

```python
import pandas as pd

excel_file_path = r"d:\Users\A.Fallah\Desktop\Price Estimation Model\Last Round\GNAF_Estimation_New.csv"
df_allestimation = pd.read_csv(excel_file_path)
df_allestimation['postal_code'] = df_allestimation['postal_code'].astype(str)
df_allestimation['Tours'] = df_allestimation['postal_code'].apply(lambda x: x[:5])
df_allestimation
```

```python
a = sum(df_allestimation['estimated_price']) / sum(df_allestimation['area'])
a
```

### Summary Stats on All Estimation

```python
df = df_allestimation.copy()
grouped_df = df.groupby(['neighborhood', 'District'])
df['record_count'] = grouped_df['estimated_price_m2'].transform('count')
df['update_fee'] = df['area'] * df['estimated_price_m2']
df['weighted_average_price_m2'] = grouped_df['update_fee'].transform('sum') / (grouped_df['area'].transform('sum') * 1000000)
df['average_house_price'] = grouped_df['update_fee'].transform('sum') / (grouped_df['estimated_price_m2'].transform('count') * 1000000000)
select = ['neighborhood', 'District', 'weighted_average_price_m2', 'average_house_price', 'record_count']
df_stat = df[select]
df_stat = df_stat.drop_duplicates()
df_stat = df_stat.sort_values('weighted_average_price_m2', ascending=False)
df_stat = df_stat.reset_index()
df_stat
```

```python
df_stat.to_csv(r"d:\Users\A.Fallah\Desktop\Price Estimation Model\Last Round\Neighborhood_Stats.csv",  index=False)
```
